# Removing values such as "UNKNOWN", "ERROR"

```python
data[~data.isin(["UNKNOWN","ERROR"]).any(axis = 1)]
```

# Inspecting more of the dataset

```python
data.sample(10)
```

```python
cols = ["Price Per Unit","Total Spent"]
for col in cols:
    data[col] = data[col].astype(float)

data["Quantity"] = data["Quantity"].astype(int)
```

```python
data.dtypes
```

```python
data["Transaction Date"] = pd.to_datetime(data["Transaction Date"])
data["Transaction ID"] = data["Transaction ID"].astype(int)
```

```python
data.dtypes
```

```python
data.head(10)
```

```python
data = data.sort_values("Transaction Date").reset_index(drop = True)
```

# Exporting the cleaned dataset into a .xlsx

```python
import openpyxl
print(f"openpyxl Version --> {openpyxl.__version__}")
```

```python
data.to_excel("Cleaned_Dataset.xlsx")

from openpyxl import load_workbook

workbook = load_workbook("Cleaned_Dataset.xlsx")
sheet = workbook.active

for column_cells in sheet.columns:
    column_letter = column_cells[0].column_letter
    max_length = 0

    for cell in column_cells:
        value = cell.value
        if hasattr(value, "strftime"):
            text = value.strftime("%Y-%m-%d %H:%M:%S")
        else:
            text = "" if value is None else str(value)
        max_length = max(max_length, len(text))

    sheet.column_dimensions[column_letter].width = max_length + 2

workbook.save("Cleaned_Dataset.xlsx")
```

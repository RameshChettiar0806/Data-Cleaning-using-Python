```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import bs4

print(f"NumPy Version --> {np.__version__}")
print(f"Pandas Version --> {pd.__version__}")
print(f"Python requests Version --> {requests.__version__}")
print(f"Beautiful Soup Version --> {bs4.__version__}")

print("\nAll libraries imported successfully")
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

data = pd.read_csv("../raw-dataset/Airbnb_Open_Data.csv")
print(f"First 5 rows of the dataset: \n{data.head(5)}")
```

```python
print(f"Shape: {data.shape}")
print(f"\nColumns: {data.columns}")

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

display(data.sample(3))
```

```python
new_columns = []
for i in data.columns:
    new_columns.append(i.lower().capitalize())

data.columns = new_columns
print(data.columns)
```

```python
columns_to_drop = ['Id', 'Reviews per month', 'Review rate number',
       'Calculated host listings count', 'Availability 365', 'House_rules', 'License']

data.drop(columns=columns_to_drop, inplace=True)
print(data.shape)
```

```python
data.drop_duplicates(inplace=True)
print(int(data.duplicated().sum()))
```

```python
data.dropna(inplace=True)
print(data.isna().sum())
print(data.shape)
```

```python
cols = ["Host_identity_verified",  "Cancellation_policy"]

for col in cols:
    data[col] = data[col].astype(str).str.upper()


display(data[cols].head(10))
```

```python
data.reset_index(inplace=True)
display(data.head(10))
```

```python
data["Price"] = data["Price"].astype(str).str.replace("$", "", regex=False).str.replace(",", "", regex=False).str.strip()
data["Service fee"] = data["Service fee"].astype(str).str.replace("$", "", regex=False).str.strip()

data["Price"] = data["Price"].astype(float)
data["Service fee"] = data["Service fee"].astype(float)

display(data[["Price", "Service fee"]].head())
```

```python
data["Last review"] = data["Last review"].astype(str).str.replace("/", "-", regex=False)
data = data.rename(columns={"index": "Original index"})
display(data.head(10))
```

# Removing Redundant Columns

```python
columns_to_keep = ['name','online_order','book_table','rate','location','approx_cost(for two people)']
data = data[columns_to_keep]
data.shape
```

# Renaming columns

```python
columns_to_rename = []
for i in data.columns:
    columns_to_rename.append(i.capitalize())

columns_to_rename
```

# Dropping Duplicates

```python
data.duplicated().value_counts()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().value_counts()
```

```python
data.shape
```

# Cleaning Individual Columns

```python
data.sample(15)
```

```python
data.columns = columns_to_rename
```

```python
data.sample(15)
```

```python
data["Online_order"] = data["Online_order"].apply(lambda x: 1 if x == "Yes" else 0)
data["Book_table"] = data["Book_table"].apply(lambda x: 1 if x == "Yes" else 0)

data.head(5)
```

```python
data["Rate"].unique()
```

```python
data["Rate"] = data["Rate"].replace(["NEW", "-"], np.nan)
```

```python
data["Rate"].unique()
```

```python
data["Location"].unique()
```

```python
data["Approx_cost(for two people)"].unique()
```

```python
data["Rate"] = data["Rate"].str.replace("/", " out of ")
```

```python
data.sample(5)
```

```python
data.reset_index()
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data
```

# Removing NaN value

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.dropna(inplace = True)
```

```python
data.shape
```

```python
data.info()
```

```python
data["Approx_cost(for two people)"] = data["Approx_cost(for two people)"].astype(str).str.replace(",","")
data["Approx_cost(for two people)"] = data["Approx_cost(for two people)"].astype(int)
```

```python
data.sample(10)
```
